## Outlier Detection
Let's identify outliers in numerical variables that might affect model performance.

```python
# Outlier detection using IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Box plots for outlier visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

outlier_summary = {}

for i, col in enumerate(numerical_cols):
    # Box plot
    df.boxplot(column=col, ax=axes[i])
    axes[i].set_title(f'Box Plot of {col}')
    axes[i].set_ylabel(col)
    
    # Detect outliers
    outliers, lower, upper = detect_outliers_iqr(df, col)
    outlier_summary[col] = {
        'count': len(outliers),
        'percentage': (len(outliers) / len(df.dropna(subset=[col]))) * 100,
        'lower_bound': lower,
        'upper_bound': upper
    }

plt.tight_layout()
plt.show()

# Print outlier summary
print("Outlier Detection Summary (IQR Method):")
print("-" * 60)
for col, info in outlier_summary.items():
    print(f"{col}:")
    print(f"  â€¢ Outliers: {info['count']} ({info['percentage']:.1f}%)")
    print(f"  â€¢ Bounds: [{info['lower_bound']:.2f}, {info['upper_bound']:.2f}]")
    print()
```

```python
# 3. Feature Importance Heatmap and Advanced Visualizations
print("ðŸ”¥ FEATURE IMPORTANCE & ADVANCED ANALYSIS")
print("=" * 60)

# Create encoded versions of categorical variables for correlation analysis
df_encoded = df.copy()

# Encode categorical variables
categorical_to_encode = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Deck', 'Side']
for col in categorical_to_encode:
    if col in df_encoded.columns:
        if df_encoded[col].dtype == 'bool':
            df_encoded[col] = df_encoded[col].astype(int)
        else:
            # Label encode for correlation analysis
            from sklearn.preprocessing import LabelEncoder
            le = LabelEncoder()
            df_encoded[col + '_encoded'] = le.fit_transform(df_encoded[col].astype(str))

# Select features for correlation analysis
correlation_features = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'TotalSpending']
correlation_features += [col + '_encoded' for col in categorical_to_encode if col + '_encoded' in df_encoded.columns]
correlation_features += ['Transported']

# Create correlation matrix
corr_data = df_encoded[correlation_features].copy()
corr_data['Transported'] = corr_data['Transported'].astype(int)

correlation_matrix = corr_data.corr()

# Create heatmap
plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', center=0,
            square=True, fmt='.3f', cbar_kws={"shrink": .8},
            mask=mask)
plt.title('Feature Correlation Heatmap with Target Variable', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Feature importance using Random Forest
print("\nðŸŒ² RANDOM FOREST FEATURE IMPORTANCE:")
print("-" * 40)

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Prepare data for Random Forest
rf_data = df.copy()
feature_names = []

# Encode categorical variables
for col in categorical_to_encode:
    if col in rf_data.columns:
        if rf_data[col].dtype == 'bool':
            rf_data[col] = rf_data[col].astype(int)
            feature_names.append(col)
        else:
            le = LabelEncoder()
            rf_data[col + '_rf'] = le.fit_transform(rf_data[col].astype(str))
            feature_names.append(col + '_rf')

# Add numerical features
feature_names += ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'TotalSpending']

# Prepare features and target
X_rf = rf_data[feature_names].fillna(0)
y_rf = rf_data['Transported'].astype(int)

# Train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_rf, y_rf)

# Get feature importance
feature_importance = pd.DataFrame({
    'feature': feature_names,
    'importance': rf.feature_importances_
}).sort_values('importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(12, 8))
sns.barplot(data=feature_importance.head(15), x='importance', y='feature', palette='viridis')
plt.title('Top 15 Feature Importance (Random Forest)', fontsize=16, fontweight='bold')
plt.xlabel('Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.tight_layout()
plt.show()

print(feature_importance.head(10))
```

```python
# 4. Advanced Interaction Analysis
print("ðŸ”„ FEATURE INTERACTION ANALYSIS")
print("=" * 60)

# Select key features for pair plot
key_features = ['Age', 'TotalSpending', 'RoomService', 'Spa', 'VRDeck']
pair_data = df[key_features + ['Transported']].copy()

# Create pair plot
plt.figure(figsize=(14, 12))
g = sns.pairplot(pair_data, hue='Transported', diag_kind='hist', 
                plot_kws={'alpha': 0.6}, diag_kws={'alpha': 0.7})
g.fig.suptitle('Pair Plot of Key Features by Transportation Status', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()

# Spending pattern analysis
print("\nðŸ’° SPENDING PATTERN ANALYSIS:")
print("-" * 40)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Spending by Age Groups and Transportation
df['AgeGroup_detailed'] = pd.cut(df['Age'], bins=[0, 18, 30, 50, 80], 
                                labels=['Young', 'Adult', 'Middle-aged', 'Senior'])

age_spending_transport = df.groupby(['AgeGroup_detailed', 'Transported'])['TotalSpending'].mean().unstack()
age_spending_transport.plot(kind='bar', ax=axes[0,0], color=['lightcoral', 'lightblue'])
axes[0,0].set_title('Average Spending by Age Group and Transportation Status')
axes[0,0].set_xlabel('Age Group')
axes[0,0].set_ylabel('Average Total Spending')
axes[0,0].legend(['Not Transported', 'Transported'])
axes[0,0].tick_params(axis='x', rotation=45)

# 2. VIP vs Regular spending patterns
vip_transport_spending = df.groupby(['VIP', 'Transported'])['TotalSpending'].mean().unstack()
vip_transport_spending.plot(kind='bar', ax=axes[0,1], color=['lightcoral', 'lightblue'])
axes[0,1].set_title('Average Spending: VIP vs Regular by Transportation Status')
axes[0,1].set_xlabel('VIP Status')
axes[0,1].set_ylabel('Average Total Spending')
axes[0,1].legend(['Not Transported', 'Transported'])
axes[0,1].tick_params(axis='x', rotation=0)

# 3. Home Planet and Destination combinations
planet_dest_transport = pd.crosstab([df['HomePlanet'], df['Destination']], 
                                   df['Transported'], normalize='index')
planet_dest_transport[True].plot(kind='bar', ax=axes[1,0], color='skyblue')
axes[1,0].set_title('Transportation Rate by Planet-Destination Combination')
axes[1,0].set_xlabel('Home Planet - Destination')
axes[1,0].set_ylabel('Transportation Rate')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. CryoSleep and Spending contradiction analysis
cryo_spending_bins = pd.cut(df['TotalSpending'], bins=[0, 1, 100, 1000, float('inf')], 
                           labels=['Zero', 'Low', 'Medium', 'High'])
cryo_contradiction = pd.crosstab([df['CryoSleep'], cryo_spending_bins], 
                                df['Transported'], normalize='index')
cryo_contradiction[True].plot(kind='bar', ax=axes[1,1], color='lightgreen')
axes[1,1].set_title('Transportation Rate by CryoSleep-Spending Pattern')
axes[1,1].set_xlabel('CryoSleep Status - Spending Level')
axes[1,1].set_ylabel('Transportation Rate')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Print interaction insights
print("\n KEY INTERACTION INSIGHTS:")
print("-" * 35)

print("1. Age Group & Transportation:")
for age_group in age_spending_transport.index:
    not_transported = age_spending_transport.loc[age_group, False]
    transported = age_spending_transport.loc[age_group, True]
    print(f"   â€¢ {age_group}: Not Transported ${not_transported:.0f}, Transported ${transported:.0f}")

print("\n2. VIP Status & Transportation:")
for vip_status in vip_transport_spending.index:
    not_transported = vip_transport_spending.loc[vip_status, False]
    transported = vip_transport_spending.loc[vip_status, True]
    vip_text = "VIP" if vip_status else "Regular"
    print(f"   â€¢ {vip_text}: Not Transported ${not_transported:.0f}, Transported ${transported:.0f}")

print(f"\n3. CryoSleep Logic Check:")
cryo_with_spending = df[(df['CryoSleep'] == True) & (df['TotalSpending'] > 0)]
print(f"   â€¢ Passengers in CryoSleep with spending: {len(cryo_with_spending)} ({len(cryo_with_spending)/len(df)*100:.1f}%)")
print(f"   â€¢ Their average transportation rate: {cryo_with_spending['Transported'].mean():.3f}")

print(f"\n4. Top Planet-Destination Combinations by Transportation Rate:")
top_combinations = planet_dest_transport[True].sort_values(ascending=False).head(5)
for combination, rate in top_combinations.items():
    print(f"   â€¢ {combination[0]} â†’ {combination[1]}: {rate:.3f}")
```

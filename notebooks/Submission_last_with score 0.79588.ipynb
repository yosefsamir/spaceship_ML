```python
def split_cabin(df):
    df[['Cabin_Deck', 'Cabin_Num', 'Cabin_Side']] = df['Cabin'].str.split('/', expand=True)
    return df.drop(columns=['Cabin'])

df_train = split_cabin(df_train)
df_test = split_cabin(df_test)

spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
```

```python
#Handling Missing Values

for col in ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Cabin_Deck', 'Cabin_Side']:
    mode = df_train[col].mode()[0]
    df_train[col].fillna(mode, inplace=True)
    df_test[col].fillna(mode, inplace=True)

for col in ['Age'] + spending_cols :
    median = df_train[col].median()
    df_train[col].fillna(median, inplace=True)
    df_test[col].fillna(median, inplace=True)
```

```python
#Handling Outliers

for col in spending_cols:
    cap = df_train[col].quantile(0.99)
    df_train[col] = df_train[col].clip(upper=cap)
    df_test[col] = df_test[col].clip(upper=cap)
```

```python
#Feature Engineering

df_train['Total_Spending'] = df_train[spending_cols].sum(axis=1)
df_test['Total_Spending'] = df_test[spending_cols].sum(axis=1)

df_train['Spending_per_Age'] = df_train['Total_Spending'] / (df_train['Age'] + 1)
df_test['Spending_per_Age'] = df_test['Total_Spending'] / (df_test['Age'] + 1)

df_train['NoSpendingWhileCryo'] = ((df_train['CryoSleep'] == True) & (df_train['Total_Spending'] == 0)).astype(int)
df_test['NoSpendingWhileCryo'] = ((df_test['CryoSleep'] == True) & (df_test['Total_Spending'] == 0)).astype(int)

df_train['Deck_Destination'] = df_train['Cabin_Deck'] + '_' + df_train['Destination']
df_test['Deck_Destination'] = df_test['Cabin_Deck'] + '_' + df_test['Destination']

df_train['Age_VIP'] = df_train['Age'] * df_train['VIP']
df_test['Age_VIP'] = df_test['Age'] * df_test['VIP']

df_train['Spending_Per_Age'] = df_train['Total_Spending'] / (df_train['Age'] + 1)
df_test['Spending_Per_Age'] = df_test['Total_Spending'] / (df_test['Age'] + 1)

df_train["Regular"] = df_train["FoodCourt"] + df_train["ShoppingMall"]
df_test["Regular"] = df_test["FoodCourt"] + df_test["ShoppingMall"]

df_train["Luxury"] = df_train["RoomService"] + df_train["Spa"] + df_train["VRDeck"]
df_test["Luxury"] = df_test["RoomService"] + df_test["Spa"] + df_test["VRDeck"]
```

```python
bool_cols = ['CryoSleep', 'VIP']
for col in bool_cols:
    df_train[col] = df_train[col].astype(bool).astype(int)
    df_test[col] = df_test[col].astype(bool).astype(int)

df_train['Transported'] = df_train['Transported'].astype(bool).astype(int)

categorical_cols = ['HomePlanet', 'Destination', 'Cabin_Deck', 'Cabin_Side', 'Deck_Destination']
df_train = pd.get_dummies(df_train, columns=categorical_cols, drop_first=True)
df_test = pd.get_dummies(df_test, columns=categorical_cols, drop_first=True)
```
